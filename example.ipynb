### Get Breathing Rate from a sample video file

```python
video_id = 'dog006'
video_path = '/home/mrahman7/Documents/' + video_id + '.mp4'  # change this to the video you want to analyze
flow_path = '/home/mrahman7/Documents/of_out/dog006'  # change this to a folder name of your choice (will be generated)
signal_collection_folder = "outputs/"

fx, fy = 0.5, 0.5  # your choice  fx=0.5 will reduce the width of the video by half (it fastens the OF calculation)
```

```python
get_gma_mv_result(video_path, flow_path, fx=fx, fy=fy)  # OF files will be written in the flow_path folder
```

```python
#%matplotlib inline
%matplotlib tk

what_to_collect = 'displacement' #'displacement' for breathing rate

collect_signals(video_id,
                video_path,
                flow_path,
                what_to_collect,
                fx=fx, fy=fy, crop=None, start_sec=0, end_sec=None, set_id=4,
                apply_mv=True, vis_on=True, graph_on=True, radius=2, output_folder=signal_collection_folder)
```

```python
%matplotlib inline

est_time, est_br, err, err_per = get_br(video_id, 4, data_folder=signal_collection_folder, verbose=True, GT_folder='GT/')

print('RMSE error: {:.2f} Standard Deviation: {:.2f}'.format(calc_rmse(np.array(err)), calc_sd(np.abs(err))))
print('Relative error MeRate: {:.2f}%'.format(np.mean(np.abs(err_per)) * 100))
print('(Percentage) RMSE error: {:.2f}% Standard Deviation: {:.2f}%'.format(calc_rmse(np.array(err_per)) * 100, calc_sd(np.array(err_per))*100))
```
